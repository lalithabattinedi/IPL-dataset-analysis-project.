11. Which batsman played the most balls?
Output: V Kohli


```python
playercount=data.groupby("batsman")["ball"].count() #will get the count
maxcount=playercount.max()#will get the max value
name=playercount.idxmax()#will get the max value's index name=player name
print(maxcount)
print(name)
name2=playercount.idxmin()#will get min values index name=lowest player
print(name2)
final2=playercount.min()#will get min value
print(final2)
```
